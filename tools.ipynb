```python
from PIL import Image

# 이미지 파일 경로 리스트
image_files = [
    "C:/Users/206539/Pictures/Screenshots/스크린샷 2024-08-08 113302.png",
    "C:/Users/206539/Pictures/Screenshots/스크린샷 2024-08-08 113412.png",
    "C:/Users/206539/Pictures/Screenshots/스크린샷 2024-08-08 113447.png",
    "C:/Users/206539/Pictures/Screenshots/스크린샷 2024-08-08 113517.png",
    "C:/Users/206539/Pictures/Screenshots/스크린샷 2024-08-08 113535.png"
]

# 이미지 객체 리스트
images = [Image.open(image) for image in image_files]

# 가로 너비와 전체 높이를 계산
width = max(image.width for image in images)
total_height = sum(image.height for image in images)

# 새로운 빈 이미지 생성
new_image = Image.new('RGB', (width, total_height))

# 각 이미지를 새로운 이미지에 붙여넣기
y_offset = 0
for image in images:
    new_image.paste(image, (0, y_offset))
    y_offset += image.height

# 결과 이미지 저장
new_image.save("C:/Users/206539/Pictures/Screenshots/Merged.png")
new_image.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# 데이터 생성
x = np.linspace(-2 * np.pi, 2 * np.pi, 200)
y_cos = np.cos(x)

# 각 값에 대해 소수점 2자리로 포매팅
formatted_cos = [f'{np.degrees(val):.2f}' for val in y_cos]

print(f'x={x}, cos(x) = {formatted_cos}')

# 그래프 그리기
plt.figure(figsize=(8, 6))
plt.plot(x, y_cos, label='cos(x)')
plt.scatter(x, y_cos)
# x축 눈금 설정
plt.xticks(
    [-2*np.pi, -3*np.pi/2, -np.pi, -np.pi/2, 0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
    [r'$-2\pi$', r'$-\frac{3\pi}{2}$', r'$-\pi$', r'$-\frac{\pi}{2}$', r'$0$', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$']
)

plt.title('Cosine Graph')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from pptx import Presentation
from pptx.util import Inches
import os

# PPT 파일 열기
ppt = Presentation()

# 이미지 파일 경로
image_folder = r"F:\2024\dev\planGA\0825\1708"
image_files = [f for f in os.listdir(image_folder) if f.endswith('.png')]

# 슬라이드 생성
slide_layout = ppt.slide_layouts[5]  # 빈 슬라이드 레이아웃 선택
slide = ppt.slides.add_slide(slide_layout)

# 이미지 배치 설정
rows = 2
cols = 6
img_width = Inches(1.5)
img_height = Inches(1.5)
spacing_x = Inches(0.2)
spacing_y = Inches(0.2)

# 이미지를 슬라이드에 추가
for i, img_file in enumerate(image_files):
    img_path = os.path.join(image_folder, img_file)
    col = i % cols
    row = i // cols
    
    left = col * (img_width + spacing_x) + Inches(0.5)
    top = row * (img_height + spacing_y) + Inches(1)
    
    slide.shapes.add_picture(img_path, left, top, width=img_width, height=img_height)

# 저장 또는 현재 열려있는 PPT에 적용
ppt.save('F:\\2024\\논문작성\\1708.pptx')
```
