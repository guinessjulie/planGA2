## Gradient-based Algorithm

```python
import numpy as np
import matplotlib.pyplot as plt

# 함수 정의
def f(x):
    return x**2

# 그래디언트(미분) 정의
def gradient(x):
    return 2*x

# 초기값 설정
x = 10
learning_rate = 0.1
tolerance = 1e-6
max_iterations = 10
print(f'iter;x     ; gradient(x)')
x_values=[x]
# 경사 하강법 반복
for i in range(max_iterations):
    grad = gradient(x)
    new_x = x - learning_rate * grad
    x_values.append((new_x))
    print(f'{i}; {x}; {grad};{new_x}')
    if abs(new_x - x) < tolerance:
        break
    x = new_x

# 그래프 시각화
x_range = np.linspace(-11, 11, 400)
y_range = f(x_range)

plt.figure(figsize=(10, 6))
plt.plot(x_range, y_range, label='f(x) = x^2')
plt.scatter(x_values, [f(x) for x in x_values], color='red', zorder=5)
for i in range(len(x_values) - 1):
    plt.arrow(x_values[i], f(x_values[i]), 
              x_values[i+1] - x_values[i], f(x_values[i+1]) - f(x_values[i]), 
              head_width=0.3, head_length=0.3, fc='red', ec='red')

plt.title('Gradient Descent Optimization')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
print(f"Minimum value of f(x) is at x = {x}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, BoundaryNorm

def color_cells_by_value(grid, filename, text=None):
    # 각 정수 값에 대응하는 RGB 색상 정의
    colors = np.array([(1, 1, 1),  # -1: 검정
                       (1, 1, 1),  # 0 : 흰색
                       (1, 0, 0),  # 빨간색
                       (0, 1, 0),  # 초록색
                       (0, 0, 1),  # 파란색
                       (1, 1, 0),  # 노란색
                       (0, 1, 1)]) # 청록색

    cmap = ListedColormap(colors)

    # 데이터 값의 범위에 따른 경계값 설정 (여기서는 -1에서 5까지)
    boundaries = [-1.5, -0.5, 0.5, 1.5, 2.5, 3.5, 4.5, 5.5]
    norm = BoundaryNorm(boundaries, cmap.N)

    # matshow로 데이터 시각화
    fig, ax = plt.subplots()
    cax = ax.matshow(grid, cmap=cmap, norm=norm)
    colorbar = fig.colorbar(cax, ticks=[-1, 0, 1, 2, 3, 4, 5])
    colorbar.set_ticklabels(['OUT', 'OUT', '1', '2', '3', '4', '5'])

    # 하단 여백을 조정하여 텍스트를 추가할 공간 확보
    plt.subplots_adjust(bottom=0.2)

    # 하단에 텍스트 추가
    if text:
        fig.text(0.5, 0.05, text, ha='center', fontsize=12)

    plt.savefig(filename)
    print(f'{filename} with {text} saved')
    plt.show()

# 테스트용 데이터
grid = np.array([
    [1, 2, 3, 4, 5],
    [0, -1, 1, 2, 3],
    [4, 5, 0, 1, 2]
])

# 함수 호출
color_cells_by_value(grid, 'test_plot.png', text='This is an example of a grid with colored cells.')
```
