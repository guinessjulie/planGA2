## Gradient-based Algorithm

```python
import numpy as np
import matplotlib.pyplot as plt

# 함수 정의
def f(x):
    return x**2

# 그래디언트(미분) 정의
def gradient(x):
    return 2*x

# 초기값 설정
x = 10
learning_rate = 0.1
tolerance = 1e-6
max_iterations = 10
print(f'iter;x     ; gradient(x)')
x_values=[x]
# 경사 하강법 반복
for i in range(max_iterations):
    grad = gradient(x)
    new_x = x - learning_rate * grad
    x_values.append((new_x))
    print(f'{i}; {x}; {grad};{new_x}')
    if abs(new_x - x) < tolerance:
        break
    x = new_x

# 그래프 시각화
x_range = np.linspace(-11, 11, 400)
y_range = f(x_range)

plt.figure(figsize=(10, 6))
plt.plot(x_range, y_range, label='f(x) = x^2')
plt.scatter(x_values, [f(x) for x in x_values], color='red', zorder=5)
for i in range(len(x_values) - 1):
    plt.arrow(x_values[i], f(x_values[i]), 
              x_values[i+1] - x_values[i], f(x_values[i+1]) - f(x_values[i]), 
              head_width=0.3, head_length=0.3, fc='red', ec='red')

plt.title('Gradient Descent Optimization')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
print(f"Minimum value of f(x) is at x = {x}")
```
